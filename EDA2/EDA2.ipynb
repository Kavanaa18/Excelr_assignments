## 2. Encoding Techniques:

```python
object_columns = df.select_dtypes(include=['object']).columns

for col in object_columns:
    unique_values = df[col].nunique()
#     value_counts = df[col].value_counts()
    print(f"{col}: {unique_values}\n")
    unique_values


# unique_values = df['workclass'].unique()
# unique_values
```

```python
df.head()
```

```python
#One-hot encoding

df_encoded = pd.get_dummies(df,columns=['sex','income'])

df_encoded.head()
```

```python
from sklearn.preprocessing import LabelEncoder

df1 = df.copy()
columns_to_encode = ["workclass","education","marital_status","occupation","relationship","race","native_country"]
label_encoder = LabelEncoder()
for col in columns_to_encode:
    df1[col + '_encoded'] = label_encoder.fit_transform(df[col])
    
df1.head()
```

```python
df.head()
```

one-hot:

pros:
* compatible with most algo
* doen't assume columns hve oridinal relation

cons:
* creates extra column which might lead to memory usage when larger categories are there

label-encoder:

pros:
* simple to understand 
* doesn't create extra columns 

cons:
* some algo may be biased to the larger label
